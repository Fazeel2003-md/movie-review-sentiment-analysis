```python
df=df.drop_duplicates()
```

```python
df['sentiment'].value_counts()
```

```python
import re
df=df.copy()
def html_tag_removal(text):
    return re.sub(r'<.*?>','',text)
df['review']=df['review'].apply(html_tag_removal)
```

```python
def url_removal(text):
    return re.sub(r'http\S+|www\S+','',text)
df['review']=df['review'].apply(url_removal)
```

```python
import string 
def punctuation_removal(text):
    return text.translate(str.maketrans('','',string.punctuation))
df['review']=df['review'].apply(punctuation_removal)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
df['sentiment']=encoder.fit_transform(df['sentiment'])
```

```python
df
```

```python
from keras.preprocessing.text import Tokenizer
tokenizer=Tokenizer(num_words=5000,oov_token='Nothing')
tokenizer.fit_on_texts(df['review'])
```

```python
df
```

```python
tokenizer.word_index
```

```python
tokenizer.word_counts
```

```python
sequences=tokenizer.texts_to_sequences(df['review'])
```

```python
from keras.utils import pad_sequences
sequences=pad_sequences(sequences,maxlen=39)
```

```python
from keras.layers import SimpleRNN,Input,Dense,Embedding,Dropout
from keras.models import Sequential
x=sequences
y=df['sentiment'].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
model = Sequential()
model.add(SimpleRNN(32, input_shape=(39,1), return_sequences=False))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])

model.fit(x_train, y_train, epochs=5,validation_data=(x_test, y_test),batch_size=64)
```

```python
model.summary()
```

```python
vocab_size=len(tokenizer.word_index)+1
```

```python
from keras.models import Sequential
from keras.layers import Embedding,LSTM, Dense, Dropout
from keras.callbacks import EarlyStopping
from keras.regularizers import l2

final_model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=64, input_length=39, mask_zero=True),
    LSTM(64,dropout=0.2, recurrent_dropout=0.2),
    Dropout(0.5),
    Dense(3, activation='softmax', kernel_regularizer=l2(0.001))
])

final_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=2,
    restore_best_weights=True
)

final_model.fit(
    x_train,
    y_train,
    epochs=10,
    validation_data=(x_test, y_test),
    batch_size=64,
    callbacks=[early_stop]
)
```

```python
final_model.evaluate(x_test, y_test)
```

```python
y_pred = final_model.predict(x_test)
y_pred_classes = y_pred.argmax(axis=1)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred_classes)
```
